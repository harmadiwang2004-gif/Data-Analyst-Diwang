**ML Model**

```python
from re import X
#add Feature list for clustering
Feature = ['GR','ILD_log10','DeltaPHI','PHIND','PE']

#Filter data model
data = data [Feature + ['Depth']].dropna ()

X = data [Feature]

data.head ()

```

```python
#Preprocessing - Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

```python
#Model Training

n_rocktype = 4 #jumlah clasifikasi yg kita inginkan

#KMeans
from sklearn.cluster import KMeans
KMeans=KMeans(n_clusters=n_rocktype, random_state=42, n_init=20)
label_KMeans = KMeans.fit_predict (X_standard)
data['Kmeans'] =label_KMeans

#GM
from sklearn.mixture import GaussianMixture
GaussianMixture = GaussianMixture(n_components=n_rocktype,covariance_type='full', random_state=42)
label_GM = GaussianMixture.fit_predict (X_standard)
data['GM']=label_GM

#AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering # Changed import
AC_model = AgglomerativeClustering(n_clusters=n_rocktype, linkage="ward") # Using AgglomerativeClustering
label_AC = AC_model.fit_predict(X_standard)
data['AC'] = label_AC


data.head()
```

```python
#ML Evaluation
from sklearn.metrics import silhouette_score

SL_Kmeans = silhouette_score(X_standard, label_KMeans)
SL_GM = silhouette_score(X_standard, label_GM)
SL_AC = silhouette_score(X_standard, label_AC)

results = pd.DataFrame({
"Model": ["Kmeans", "GM", "AC"],
"Silhouette": [SL_Kmeans, SL_GM, SL_AC]
})
print(results)
```

```python
sns.scatterplot(data=data, x=GR, y=DeltaPHI, palette="tab20", hue=data['Kmeans'])
plt.title('GR vs DeltaPHI After ML-Kmeans')
plt.show()
```

```python
sns.scatterplot(data=data, x=GR, y=DeltaPHI, palette="tab20", hue=data['GM'])
plt.title('GR vs DeltaPHI After ML-Gausian')
plt.show()
```

```python
sns.scatterplot(data=data, x=GR, y=DeltaPHI, palette="tab20", hue=data['AC'])
plt.title('GR vs DeltaPHI After ML-AgglomerativeClustering')
plt.show()
```

Hasil analisa dari ketiga ML yaitu Kmeans, GM dan AC

1. Hasil Analisa ML Kmeans pada crosplot GR-DeltaPHI

Makna sumbu pada ML Kmeans :
- Sumbu X = GR (Gamma Ray)  indikator kandungan shale. Makin besar, makin shaly.

- Sumbu Y = DeltaPHI  selisih porositas (biasanya density-neutron).

- Nilai positif sering diasosiasikan dengan gas effect atau porositas efektif.

- Nilai negatif bisa mengindikasikan tight zone, shale, atau efek densitas tinggi.

- hasil clustering K-Means pada crossplot GR-DeltaPHI menunjukkan empat kelompok utama yang merepresentasikan variasi karakter log.
A. Cluster dengan GR rendah dan DeltaPHI mendekati nol hingga positif mengindikasikan zona reservoir yang relatif bersih dan berporositas baik,
B. Cluster dengan GR tinggi didominasi oleh zona shale. Overlap antar cluster menunjukkan heterogenitas formasi dan keterbatasan metode unsupervised dalam memisahkan litologi tanpa data pembanding.

2. Hasil Analisa ML GM pada crosplot GR-DeltaPHI

Makna sumbu pada ML GM :

- GR (X-axis)
 Indikator kandungan shale
 Nilai tinggi = shale-rich

- DeltaPHI (Y-axis)
 Selisih porositas (densityneutron)
 Positif: indikasi porositas efektif / gas effect
 Negatif: tight zone, shale, atau mineral berat

- Data GR-DeltaPHI tidak terpisah secara tegas

- GMM memberikan segmentasi yang lebih geologis

- Cluster dengan DeltaPHI positif dan GR menengah paling potensial sebagai reservoir

- Hasil clustering Gaussian menunjukkan bahwa karakter reservoir dan non-reservoir pada data GR- deltaPHI memiliki overlap yang signifikan, sehingga pendekatan probabilistik lebih representatif dibanding metode clustering berbasis jarak.

3. Hasil Analisa ML AC pada crosplot GR-DeltaPHI

Makna sumbu pada ML AC:

- Sumbu X (GR)
 indikator shale content

- Sumbu Y (DeltaPHI)
 karakter porositas dan efek fluida

Warna berbeda = hasil pemotongan dendrogram pada jumlah cluster tertentu, bukan hasil optimasi otomatis.

- Hasil clustering AC (Agglomerative Clustering) memberikan segmentasi berbasis kemiripan bertahap yang cukup representatif untuk data GRDeltaPHI, namun tetap memerlukan kalibrasi geologi agar interpretasi tidak bersifat subjektif.


**"Kesimpulan"**

1. K-Means
Cocok untuk eksplorasi awal dan data sederhana, namun kurang merepresentasikan transisi litologi nyata.

2. GMM
Paling representatif secara petrofisika, mampu menangkap overlap dan ketidakpastian data.

3. Agglomerative
Memberikan struktur hierarkis yang baik, namun hasil sangat bergantung pada pemilihan threshold.

KMeans memberikan hasil clustering yang paling jelas dan mudah diinterpretasikan secara visual pada crossplot GRPHI. Namun, metode ini cenderung menyederhanakan heterogenitas data dan kurang merepresentasikan overlap litologi yang nyata, dibandingkan GMM yang lebih sesuai untuk data petrofisika yang bersifat kontinu dan tidak terpisah tegas.
